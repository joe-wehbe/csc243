Question 1

```python
matrix = []
students_number = int(input("Enter the number of students taking the exam: "))
questions_number = int(input("Enter the number of questions of the exam: "))
key = input("Enter the correct answers of the exam (key)(separated by commas):").split(',')
while len(key) != questions_number:
    key = input("Key should be equal to the number of questions. Enter the correct answers of the exam (key): ").split(',')
    print("The correct answers of the exam are:",key)
for rows in range(students_number):
    matrix.append([])
    print("Enter the answers of the student",rows + 1,"(separated by commas):")
    answers = input().split(',')
    for columns in range(questions_number):
        matrix[rows].append(answers[columns]) 
print("The",rows + 1,"students' answers are:",matrix)
scores_list = []
for row in range(len(matrix)): 
    score = 0
    for col in range(len(matrix[row])): 
        if matrix[row][col] == key[col]:
            score += 100 / questions_number
    scores_list.append(score)
for x in range(len(scores_list)-1):
    for y in range(x + 1,len(scores_list)):
        if scores_list[x] < scores_list[y]:
            temp = scores_list[x]
            scores_list[x] = scores_list[y]
            scores_list[y] = temp
total = 0
for r in range(students_number):
    print("Student ranked number",r+1,"scored:",scores_list[r])
    total += scores_list[r]
average = total / students_number
print("Average is:",average)
print("Maximum is:",max(scores_list))
print("Minimum is:",min(scores_list))
```
