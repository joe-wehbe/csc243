```python
items = int(input("Enter the number of items to add to the dictionary: "))
dictionary = {}
for number in range(items):
    print("Enter key",number + 1,"of the dictionary: ")
    key = input()
    print("Enter the value of key",number + 1,"in the dictionary: ")
    value = input()
    dictionary[key] = value
print("The dictionary is",dictionary)
invert = {}
for key in dictionary:
    value = dictionary[key]
    if value in invert:
        invert[value].append(key)
    else:
        invert[value] = [key]
print("The invert of the dictionary is",invert)
```
